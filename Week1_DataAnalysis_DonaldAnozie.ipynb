```python
# ===== WEEK 1 DATA ANALYSIS ASSIGNMENT =====
# Student: DONALD ANOZIE -DA
# Date: 07-08-2025
# Platform: Google Colab on Windows 11

print("Starting Week 1 Assignment...")

# Import the tools we need for data analysis
import pandas as pd          # For working with data tables
import numpy as np           # For math calculations
import matplotlib.pyplot as plt  # For creating charts
import seaborn as sns        # For making pretty charts

# Settings to make our charts look good
plt.style.use('default')
%matplotlib inline

print("‚úÖ Success! All tools are ready to use")
print("üíª Running on Windows 11 with Google Colab")
```

```python
# Load our dataset - we'll use Titanic passenger data
# This data contains information about people who were on the Titanic ship
print("Loading dataset...")

df = sns.load_dataset('titanic')  # 'df' stands for 'dataframe' - it's like an Excel spreadsheet
dataset_name = "Titanic Passenger Data"

print(f"‚úÖ Dataset loaded successfully!")
print(f"üìä Dataset name: {dataset_name}")
print(f"üìè Size: {df.shape[0]} rows (passengers) √ó {df.shape[1]} columns (information about each passenger)")
print(f"üíæ Data size: {df.memory_usage(deep=True).sum() / 1024:.1f} KB")
print()
print("The columns (types of information) we have:")
print(list(df.columns))
```

```python
# ====== ASSIGNMENT REQUIREMENT 1: DATASET PREVIEW ======
# We need to show the first few rows of our data using .head()

print("üîç DATASET PREVIEW - SHOWING FIRST 5 ROWS")
print("=" * 60)
print()

print(f"üìã Dataset: {dataset_name}")
print(f"üìä Total size: {df.shape[0]} passengers, {df.shape[1]} pieces of information per passenger")
print()

print("üëÄ Here are the first 5 passengers using .head():")
print("(This is like looking at the first 5 rows of an Excel spreadsheet)")
print()
df.head()  # This shows the first 5 rows
```

```python
# ====== ASSIGNMENT REQUIREMENT 2: SUMMARY STATISTICS ======
# We need to show statistical summary using .describe()

print("üìä SUMMARY STATISTICS - MATHEMATICAL SUMMARY OF OUR DATA")
print("=" * 60)
print()

print("üî¢ Statistical summary of numerical columns using .describe():")
print("(This shows averages, minimums, maximums, etc.)")
print()
df.describe()  # This creates a statistical summary
```

```python
# ====== ASSIGNMENT REQUIREMENT 3: DATA VISUALIZATION ======
# We need to create at least 1 chart. We'll create 4 different charts!

print("üìà DATA VISUALIZATIONS - CREATING CHARTS FROM OUR DATA")
print("=" * 60)
print()

# Set up our chart area (like preparing a canvas for painting)
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Titanic Dataset Analysis - 4 Different Views of Our Data', fontsize=16, fontweight='bold')

# Chart 1: How many people survived vs died (Bar Chart)
survival_counts = df['survived'].value_counts()
axes[0, 0].bar(['Died', 'Survived'], survival_counts.values,
               color=['red', 'green'], alpha=0.7)
axes[0, 0].set_title('1. Survival Count: How Many Lived vs Died')
axes[0, 0].set_ylabel('Number of People')
axes[0, 0].grid(True, alpha=0.3)

# Chart 2: Age distribution (Histogram)
ages = df['age'].dropna()  # Remove missing ages
axes[0, 1].hist(ages, bins=20, color='skyblue', alpha=0.7, edgecolor='black')
axes[0, 1].set_title('2. Age Distribution: How Old Were Passengers')
axes[0, 1].set_xlabel('Age (years)')
axes[0, 1].set_ylabel('Number of People')
axes[0, 1].grid(True, alpha=0.3)

# Chart 3: Passenger class distribution (Bar Chart)
class_counts = df['class'].value_counts()
axes[1, 0].bar(class_counts.index, class_counts.values,
               color=['gold', 'silver', '#CD7F32'], alpha=0.7)
axes[1, 0].set_title('3. Ticket Class: First, Second, or Third Class')
axes[1, 0].set_xlabel('Class (1st=Expensive, 3rd=Cheap)')
axes[1, 0].set_ylabel('Number of People')
axes[1, 0].grid(True, alpha=0.3)

# Chart 4: Ticket prices (Box Plot)
fares = df['fare'].dropna()  # Remove missing fares
box_plot = axes[1, 1].boxplot(fares, patch_artist=True)
box_plot['boxes'][0].set_facecolor('lightcoral')
box_plot['boxes'][0].set_alpha(0.7)
axes[1, 1].set_title('4. Ticket Prices: How Much Did People Pay')
axes[1, 1].set_ylabel('Price ($)')
axes[1, 1].grid(True, alpha=0.3)

# Show all our charts
plt.tight_layout()
plt.show()

print("‚úÖ All 4 visualizations created successfully!")
print("üìä Charts show: survival rates, age distribution, passenger classes, and ticket prices")
```
